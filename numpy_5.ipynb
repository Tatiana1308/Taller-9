# Crear ndarray con datos aleatorios


```python
aleatorio = np.random.rand(5,5)
aleatorio
```

# Operaciones con ndarrays


```python
suma= np.sum(aleatorio)
suma
```

```python
media= np.mean(aleatorio)
media
```

```python
mediana= np.median(aleatorio)
mediana
```

```python
desviacion= np.std(aleatorio)
desviacion
```

```python
seno = np.sin(aleatorio)
seno
```

```python
grados = np.rad2deg(aleatorio)
grados
```

```python
ordenado= np.sort(aleatorio)
ordenado
```
